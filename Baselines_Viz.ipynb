```python
from baselines.common import plot_util as pu
import matplotlib.pyplot as plt
import numpy as np
```

```python
fps = []
plt.rcParams["figure.figsize"] = (50,25)
```

```python
def plot_lc(fps_list):
    for i in range(len(fps_list)):
        plt.subplot(421+i)
        fps1 = fps_list[i]
        for fp in fps1:
            results =  pu.load_results(fp)
            r = results[0]
            plt.plot(np.cumsum(r.monitor.l), pu.smooth(r.monitor.r, radius=100),label=fp)
            plt.legend()
    plt.show()
```
