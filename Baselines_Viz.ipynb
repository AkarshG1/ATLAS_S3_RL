```python
from baselines.common import plot_util as pu
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
```

```python
def find_t_at_r(reward_thres, roll_mean_df, log_df):
    thres = roll_mean_df.loc[roll_mean_df.r >= reward_thres]
    idx = thres.index[0]
    time_series = np.cumsum(log_df.l)
    return time_series[idx]

def plot_r_thres(r):
    plt.plot([0, 1e7], [r, r], color='k', linestyle='-', linewidth=2)

def plot_t_thres(t, max_y):
    plt.plot([t, t], [0, max_y], color='k', linestyle='-', linewidth=2)
    
def find_r_at_t(time_thres , roll_mean_df, log_df):
    time_series = np.cumsum(log_df.l)
    thres = time_series.loc[time_series >= time_thres]
    idx = thres.index[0]
    return roll_mean_df.r[idx]
```

```python
def get_thres(key):
    game = re.search(r'\w{2,5}(?=_)',key).group()
    #Human level scores as declared in https://www.nature.com/articles/nature14236.pdf pg 11
    thresholds = {\
        'si' : 1652/4,\
        'da' : 3401/4,\
        'pong' : 9.3/4, \
        'qb' : 13455/4          
    }
    thres=0
    try:
        thres = thresholds[game]
    except KeyError: 
        print("key doesn't contain game!")
    return thres 
```

```python
def plot_lc(games_algo_dict):
    i = 0
    for key in games_algo_dict:
        plt.subplot(421+i)
        fps1 = games_algo_dict[key]
        time_thres=0
        t_to_thres=0
        j=0
        for fp in fps1:
            results =  pu.load_results(fp)
            r = results[0]
            m = r.monitor
            u = m.rolling(1000).mean()
            s = m.rolling(1000).std()
            s_mean = s.r[1000:].mean(0)
            reward_thres= get_thres(key)
            try:
                t_to_thres = find_t_at_r(reward_thres, u, m)
            except:
                t_to_thres = 0
            r_at_t=0
            if t_to_thres!=0:
                r_at_t = find_r_at_t(t_to_thres , u, m)
            print(t_to_thres,r_at_t,s_mean)
            #plt.plot(np.cumsum(r.monitor.l), pu.smooth(m.r, radius=1000),label=fp)
            plt.plot(np.cumsum(r.monitor.l), u.r,label=fp)
            #plt.fill_between(np.cumsum(r.monitor.l), u.r+s.r, u.r-s.r, alpha=0.25)
            plot_r_thres(reward_thres)
            if j==0:
                plot_t_thres(t_to_thres,4*reward_thres)
            plt.legend()
            j += 1
        i += 1
    plt.show()
```
